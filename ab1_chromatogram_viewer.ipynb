# AB1 Chromatogram Viewer (Jupyter)

Interactive chromatogram viewer for **Sanger `.ab1`** files:

- Reads an `.ab1` with **Biopython**
- Automatically maps **ABI channels (DATA9–DATA12)** to **A/C/G/T**
- Displays an interactive chromatogram (scroll by bases), base calls, and per-base quality bars
- Hover:
  - Over a peak trace → shows which base trace (A/C/G/T) and intensity
  - Over a base letter → shows base and its Phred quality
- Click a base letter → toggles it to **`N`** (and back to the original base)
- Buttons:
  - **TRIM** using a Chromas-like sliding-window mean (`trim_by_quality`)
  - **MASK** low-quality bases to `N` (`mask_low_quality`)
  - Export **trimmed+edited** as **FASTA** or **FASTQ** (with qualities)

**Minimal dependencies:** `biopython`, `numpy`, `plotly`, `ipywidgets`.

```python
# Cell 1 — Load an AB1 file and extract traces, basecalls, peak positions, and qualities

import numpy as np
from Bio import SeqIO

# -----------------------------
# INPUT: set your AB1 path here
# -----------------------------
AB1_PATH = "data/sample1.ab1"  # <-- change this

rec = SeqIO.read(AB1_PATH, "abi")
raw = rec.annotations["abif_raw"]

# Basecalls (PBAS2) and peak locations (PLOC1)
pbas = raw.get("PBAS2")
if isinstance(pbas, bytes):
    pbas = pbas.decode(errors="ignore")
seq = str(pbas)

ploc = raw.get("PLOC1")  # trace indices for each basecall (len == len(seq))

# Qualities: prefer Biopython's phred_quality; fallback to PCON2 if needed
quals = rec.letter_annotations.get("phred_quality", None)
if quals is None or len(quals) != len(seq):
    pcon = raw.get("PCON2")
    quals = list(pcon) if pcon is not None else [None] * len(seq)

# Raw channels: typically DATA9..DATA12 exist
data_keys = ["DATA9", "DATA10", "DATA11", "DATA12"]
channels = {k: np.array(raw[k], dtype=float) for k in data_keys if k in raw}

if len(channels) != 4:
    raise ValueError(f"Expected 4 trace channels (DATA9..DATA12). Found: {list(channels.keys())}")

trace_len = len(next(iter(channels.values())))

print("AB1 loaded:", AB1_PATH)
print("Basecalls length:", len(seq))
print("Trace length:", trace_len)
print("Available channels:", list(channels.keys()))
```

```python
# Cell 2 — Automatically assign DATA channels to A/C/G/T and build trace_by_base

import numpy as np

def auto_assign_channels_to_bases(seq, ploc, channels, bases=("A","C","G","T"), sample_cap=5000):
    """Infer which ABI trace channel corresponds to A/C/G/T.

    Strategy:
      - At each basecall peak position (PLOC1), we measure each channel's intensity.
      - For positions called as 'A', we compute the mean intensity per channel; same for C/G/T.
      - We then greedily assign a one-to-one mapping base->channel using the highest mean intensities.

    Args:
        seq: Basecalled sequence (string).
        ploc: Peak locations (list/array of ints), one per basecall.
        channels: Dict like {"DATA9": np.array([...]), ...} for 4 channels.
        bases: Tuple of bases to map (default A,C,G,T).
        sample_cap: Maximum number of basecalls to use for mapping (speed/stability).

    Returns:
        base_to_chan: dict base->channel_key (e.g., {"A":"DATA10",...})
        chan_to_base: dict channel_key->base
        scores: numpy array (4x4) of mean intensities [base,channel]
        chan_keys: list of channel keys in column order for scores
    """
    idx = [(i, b) for i, b in enumerate(seq) if b in bases]
    if len(idx) == 0:
        raise ValueError("No A/C/G/T basecalls found to infer channels.")
    if len(idx) > sample_cap:
        idx = idx[:sample_cap]

    chan_keys = list(channels.keys())
    scores = np.zeros((len(bases), len(chan_keys)), dtype=float)

    for bi, base in enumerate(bases):
        pos_list = [ploc[i] for i, b in idx
                    if b == base and 0 <= ploc[i] < len(next(iter(channels.values())))]
        if len(pos_list) == 0:
            scores[bi, :] = -1e9
            continue
        pos_arr = np.array(pos_list, dtype=int)
        for cj, ck in enumerate(chan_keys):
            scores[bi, cj] = float(np.mean(channels[ck][pos_arr]))

    # Greedy one-to-one assignment
    remaining_bases = set(range(len(bases)))
    remaining_chans = set(range(len(chan_keys)))
    base_to_chan = {}
    chan_to_base = {}

    pairs = [(scores[bi, cj], bi, cj) for bi in range(len(bases)) for cj in range(len(chan_keys))]
    pairs.sort(reverse=True, key=lambda x: x[0])

    for score, bi, cj in pairs:
        if bi in remaining_bases and cj in remaining_chans:
            base_to_chan[bases[bi]] = chan_keys[cj]
            chan_to_base[chan_keys[cj]] = bases[bi]
            remaining_bases.remove(bi)
            remaining_chans.remove(cj)
        if not remaining_bases or not remaining_chans:
            break

    if len(base_to_chan) < 4:
        raise ValueError("Could not assign all 4 bases to channels confidently.")

    return base_to_chan, chan_to_base, scores, chan_keys

base_to_chan, chan_to_base, score_matrix, chan_keys = auto_assign_channels_to_bases(seq, ploc, channels)

print("Auto channel assignment (base -> DATA key):")
for b in "ACGT":
    print(f"  {b} -> {base_to_chan[b]}")

# Build base-labeled traces for plotting (A/C/G/T)
trace_by_base = {b: channels[base_to_chan[b]] for b in "ACGT"}
```

```python
# Cell 3 — Interactive viewer (trim/mask/edit/export)

# If needed, install dependencies in the current kernel:
# import sys
# !{sys.executable} -m pip install plotly ipywidgets anywidget

import plotly.graph_objects as go
import ipywidgets as widgets
from IPython.display import display
from pathlib import Path
from typing import List
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord
from Bio import SeqIO
import numpy as np

# -----------------------------
# Quality utilities (your functions)
# -----------------------------
def mask_low_quality(seq: str, quals: List[int], min_q: int) -> str:
    """Replace bases with Q < min_q by 'N'."""
    return "".join(b if (q is not None and q >= min_q) else "N" for b, q in zip(seq, quals))

def trim_by_quality(seq: str, quals: List[int], min_q: int, window: int, min_len: int):
    """Trim low-quality ends using a sliding window mean (Chromas-like).

    Returns:
        trimmed_seq, trimmed_quals, left_idx, right_idx
    where left_idx/right_idx are 0-based positions on the original sequence (right_idx is exclusive).
    """
    if not seq or not quals or len(seq) != len(quals):
        return "", [], 0, 0

    n = len(seq)
    if n < window:
        good = [i for i, q in enumerate(quals) if q is not None and q >= min_q]
        if not good:
            return "", [], 0, 0
        left, right = good[0], good[-1] + 1
        if right - left < min_len:
            return "", [], 0, 0
        return seq[left:right], quals[left:right], left, right

    # prefix sums for window mean (treat None as 0)
    ps = [0]
    for q in quals:
        ps.append(ps[-1] + (q if q is not None else 0))

    def win_mean(i: int) -> float:
        return (ps[i + window] - ps[i]) / window

    good_windows = [i for i in range(n - window + 1) if win_mean(i) >= min_q]
    if not good_windows:
        return "", [], 0, 0

    left = good_windows[0]
    right = good_windows[-1] + window

    # tighten to first/last base >= min_q
    while left < right and (quals[left] is None or quals[left] < min_q):
        left += 1
    while right > left and (quals[right - 1] is None or quals[right - 1] < min_q):
        right -= 1

    if right - left < min_len:
        return "", [], 0, 0

    return seq[left:right], quals[left:right], left, right

def export_trimmed_edited(out_path: str, record_id: str, seq_str: str, qual_list: List[int], fmt: str):
    """Export trimmed+edited sequence to FASTA or FASTQ."""
    rec_out = SeqRecord(Seq(seq_str), id=record_id, description="trimmed+edited from AB1 viewer")
    if fmt.lower() == "fastq":
        if len(seq_str) != len(qual_list):
            raise ValueError("FASTQ export requires qualities with the same length as the sequence.")
        rec_out.letter_annotations["phred_quality"] = [int(q) if q is not None else 0 for q in qual_list]
    SeqIO.write([rec_out], out_path, fmt.lower())

# -----------------------------
# Viewer state
# -----------------------------
orig_seq = seq
orig_quals = list(quals)

edited_seq = list(orig_seq)      # editable full-length sequence
view_left = 0                    # current view on original indices
view_right = len(orig_seq)

# Trim region on original indices (export uses only this region; view remains full)
trim_left = 0
trim_right = len(orig_seq)

def current_view_seq_and_quals():
    """Return (sequence, qualities) for the current view segment."""
    return "".join(edited_seq[view_left:view_right]), orig_quals[view_left:view_right]

def get_window_by_bases_in_view(base_start_view, base_end_view, pad=40):
    """Map base window (relative to view) -> trace window using PLOC1."""
    view_len = view_right - view_left
    base_start_view = max(0, min(int(base_start_view), max(0, view_len - 1)))
    base_end_view = max(base_start_view + 1, min(int(base_end_view), view_len))

    abs_start = view_left + base_start_view
    abs_end = view_left + base_end_view

    x0 = max(0, int(ploc[abs_start]) - pad)
    x1 = min(trace_len, int(ploc[abs_end - 1]) + pad)
    return abs_start, abs_end, x0, x1

# -----------------------------
# Widgets
# -----------------------------
win_size = widgets.IntSlider(value=120, min=30, max=400, step=10, description="Bases/window")
start_slider = widgets.IntSlider(value=0, min=0, max=max(0, (view_right-view_left)-1), step=1, description="Start base")
pad_slider = widgets.IntSlider(value=40, min=0, max=200, step=10, description="Pad")

qmask_slider = widgets.IntSlider(value=20, min=0, max=40, step=1, description="Mask Q<")
trim_minq_slider = widgets.IntSlider(value=20, min=0, max=40, step=1, description="Trim minQ")
trim_window_slider = widgets.IntSlider(value=10, min=3, max=60, step=1, description="Trim window")
trim_minlen_slider = widgets.IntSlider(value=80, min=20, max=2000, step=10, description="Trim minLen")

btn_mask = widgets.Button(description="Apply MASK (N)", button_style="warning")
btn_trim = widgets.Button(description="Apply TRIM", button_style="info")
btn_reset_view = widgets.Button(description="Reset view", button_style="")
btn_reset_edits = widgets.Button(description="Reset edits", button_style="")

out_dir = Path("viewer_out")
out_dir.mkdir(exist_ok=True)
export_name = widgets.Text(value="trimmed_edited", description="Basename")
btn_export_fasta = widgets.Button(description="Export FASTA", button_style="success")
btn_export_fastq = widgets.Button(description="Export FASTQ", button_style="success")

status = widgets.HTML(value="")

# FigureWidget requires 'anywidget' with recent plotly versions.
fig = go.FigureWidget()

current_abs_base_indices = []

def update_start_slider_bounds():
    """Update slider bounds after trimming changes the view length."""
    view_len = view_right - view_left
    start_slider.max = max(0, view_len - 1)
    if start_slider.value > start_slider.max:
        start_slider.value = start_slider.max

def render():
    """Render chromatogram + basecalls + quality bars for the current window."""
    global current_abs_base_indices

    view_len = view_right - view_left
    if view_len <= 0:
        fig.data = []
        status.value = "<b>No data to display.</b>"
        return

    base_start_view = int(start_slider.value)
    base_end_view = min(view_len, base_start_view + int(win_size.value))

    abs_start, abs_end, x0, x1 = get_window_by_bases_in_view(base_start_view, base_end_view, pad=int(pad_slider.value))

    # Guard: empty/invalid trace window (prevents max() on empty slices)
    if x1 <= x0:
        with fig.batch_update():
            fig.data = []
            fig.layout = {}
        status.value = (
            f"<b>No trace data in window.</b> x0={x0}, x1={x1}. "
            "Try increasing Pad or moving Start base."
        )
        return

    x = np.arange(x0, x1)

    with fig.batch_update():
        fig.data = []

        #fig.layout = {}
        fig.layout.shapes = ()

        # Highlight trimmed-out regions (outside trim_left:trim_right) while keeping full view visible.
        shapes = []
        try:
            # Convert base indices to trace x positions using PLOC1.
            x_trim_left = int(ploc[trim_left]) if 0 <= trim_left < len(ploc) else 0
            x_trim_right = int(ploc[trim_right - 1]) if 0 < trim_right <= len(ploc) else trace_len

            # Left trimmed-out region
            if x0 < x_trim_left:
                shapes.append(dict(
                    type="rect", xref="x", yref="paper",
                    x0=x0, x1=min(x1, x_trim_left),
                    y0=0, y1=1,
                    fillcolor="yellow", opacity=0.20,
                    line=dict(width=0),
                    layer="below",
                ))

            # Right trimmed-out region
            if x1 > x_trim_right:
                shapes.append(dict(
                    type="rect", xref="x", yref="paper",
                    x0=max(x0, x_trim_right), x1=x1,
                    y0=0, y1=1,
                    fillcolor="yellow", opacity=0.20,
                    line=dict(width=0),
                    layer="below",
                ))
        except Exception:
            # If anything goes wrong, skip highlighting rather than breaking rendering.
            shapes = []

        if shapes:
            fig.update_layout(shapes=shapes)

        # Chromatogram traces
        for b in "ACGT":
            fig.add_trace(go.Scatter(
                x=x, y=trace_by_base[b][x0:x1],
                mode="lines", name=b,
                hovertemplate="Trace: %{fullData.name}<br>Trace index: %{x}<br>Intensity: %{y}<extra></extra>"
            ))

        # Basecalls shown (still filtered so we only draw letters that fall in the trace window)
        abs_inds = [i for i in range(abs_start, abs_end) if x0 <= ploc[i] < x1]
        current_abs_base_indices = abs_inds
        locs = [ploc[i] for i in abs_inds]
        bases_txt = [edited_seq[i] for i in abs_inds]
        q_txt = [orig_quals[i] for i in abs_inds]

        custom = np.column_stack([
            [q if q is not None else -1 for q in q_txt],
            abs_inds
        ])

        maxima = []
        for b in "ACGT":
            seg = trace_by_base[b][x0:x1]
            if len(seg):
                maxima.append(float(np.max(seg)))
        y_top = max(maxima) if maxima else 1.0
        y_text = y_top * 1.03 if y_top > 0 else 1.0

        fig.add_trace(go.Scatter(
            x=locs, y=[y_text]*len(locs),
            mode="text",
            text=bases_txt,
            textfont=dict(size=10),
            name="Base calls",
            hovertemplate="Base: %{text}<br>Q: %{customdata[0]}<br>Abs index: %{customdata[1]}<extra></extra>",
            customdata=custom,
        ))

        custom_bar = np.column_stack([bases_txt, abs_inds])
        fig.add_trace(go.Bar(
            x=locs,
            y=[q if q is not None else 0 for q in q_txt],
            name="Quality",
            opacity=0.5,
            yaxis="y2",
            hovertemplate="Quality<br>Base: %{customdata[0]}<br>Q: %{y}<br>Abs index: %{customdata[1]}<extra></extra>",
            customdata=custom_bar,
        ))

        # ✅ Robust X-axis labels: build ticks from abs_start..abs_end (NOT from filtered abs_inds)
        n_ticks = 10
        step = max(1, (abs_end - abs_start) // n_ticks)
        tick_base_inds = list(range(abs_start, abs_end, step))
        tickvals = [ploc[i] for i in tick_base_inds]
        ticktext = [str(i + 1) for i in tick_base_inds]  # 1-based numbering

        fig.update_layout(
            title=f"{AB1_PATH} | TRIM export abs {trim_left}-{trim_right-1} (len={trim_right-trim_left})",
            autosize=True,
            height=560,
            hovermode="closest",
            legend=dict(orientation="h"),
            margin=dict(l=40, r=20, t=50, b=40),

            # ✅ X labels as sequence positions
            xaxis=dict(
                title="Sequence position (1-based)",
                tickmode="array",
                tickvals=tickvals,
                ticktext=ticktext,
            ),

            yaxis=dict(title="Intensity"),
            yaxis2=dict(title="Phred Q", overlaying="y", side="right", range=[0, 45], showgrid=False),
        )

    status.value = (
        f"<b>View:</b> abs bases {view_left}-{view_right-1} (len={view_right-view_left}) | "
        f"<b>TRIM export:</b> abs bases {trim_left}-{trim_right-1} (len={trim_right-trim_left}) | "
        f"<b>Click a base letter</b> to toggle 'N' / original."
    )

def wire_click_handler():
    """Attach click-to-edit handler to the base-call text trace."""
    def on_base_click(trace, points, selector):
        if trace.name != "Base calls":
            return
        if not points.point_inds:
            return
        k = points.point_inds[0]
        abs_i = current_abs_base_indices[k]
        edited_seq[abs_i] = "N" if edited_seq[abs_i] != "N" else orig_seq[abs_i]
        render()
        wire_click_handler()

    for tr in fig.data:
        if tr.name == "Base calls":
            tr.on_click(on_base_click)

def on_any_slider_change(change):
    render()
    wire_click_handler()

start_slider.observe(on_any_slider_change, names="value")
win_size.observe(on_any_slider_change, names="value")
pad_slider.observe(on_any_slider_change, names="value")

# -----------------------------
# Button callbacks
# -----------------------------
def on_mask_clicked(_):
    """Apply low-quality masking (Q < threshold -> 'N') to the current view segment."""
    min_q = int(qmask_slider.value)
    s_view = "".join(edited_seq[view_left:view_right])
    q_view = orig_quals[view_left:view_right]
    s_masked = mask_low_quality(s_view, q_view, min_q=min_q)
    edited_seq[view_left:view_right] = list(s_masked)
    render()
    wire_click_handler()

def on_trim_clicked(_):
    """Compute trimming boundaries on the full read.

    This does NOT change what is displayed (the chromatogram always shows the full read).
    The trimmed-out regions are highlighted, and exports use only the untrimmed segment.
    """
    global trim_left, trim_right

    min_q = int(trim_minq_slider.value)
    window = int(trim_window_slider.value)
    min_len = int(trim_minlen_slider.value)

    s_full = "".join(edited_seq)
    t_seq, t_q, left, right = trim_by_quality(s_full, orig_quals, min_q=min_q, window=window, min_len=min_len)

    if not t_seq:
        status.value = "<b>TRIM failed:</b> no segment passed the trimming criteria."
        return

    trim_left, trim_right = left, right

    # Keep the current view; just re-render so the highlight updates
    render()
    wire_click_handler()

def on_reset_view_clicked(_):
    """Reset only the viewing window position (keeps edits and trimming)."""
    start_slider.value = 0
    #fig.layout.width = None
    #fig.layout.autosize = True
    render()
    force_relayout()
    wire_click_handler()

def on_reset_edits_clicked(_):
    """Reset all edits AND trimming (restore original basecalls and full export region)."""
    global edited_seq, trim_left, trim_right
    edited_seq = list(orig_seq)
    trim_left, trim_right = 0, len(orig_seq)
    render()
    wire_click_handler()

def on_export_fasta(_):
    """Export trimmed+edited as FASTA (uses trimming boundaries, not the current view)."""
    s_trim = "".join(edited_seq[trim_left:trim_right])
    q_trim = orig_quals[trim_left:trim_right]
    out = out_dir / f"{export_name.value}.fasta"
    export_trimmed_edited(str(out), rec.id, s_trim, q_trim, fmt="fasta")
    status.value = f"<b>Exported FASTA:</b> {out}"


def on_export_fastq(_):
    """Export trimmed+edited as FASTQ (uses trimming boundaries, not the current view)."""
    s_trim = "".join(edited_seq[trim_left:trim_right])
    q_trim = orig_quals[trim_left:trim_right]
    out = out_dir / f"{export_name.value}.fastq"
    export_trimmed_edited(str(out), rec.id, s_trim, q_trim, fmt="fastq")
    status.value = f"<b>Exported FASTQ:</b> {out}"

def force_relayout():
    """Force a Plotly relayout so the widget recomputes its width."""
    dummy = dict(
        type="rect", xref="paper", yref="paper",
        x0=0, x1=0, y0=0, y1=0,
        line=dict(width=0),
        fillcolor="rgba(0,0,0,0)",
        opacity=0,
        layer="below"
    )
    with fig.batch_update():
        # add a dummy shape then remove it -> triggers a 'strong' relayout
        shapes = list(fig.layout.shapes) if fig.layout.shapes else []
        shapes.append(dummy)
        fig.update_layout(shapes=shapes)
        shapes.pop()
        fig.update_layout(shapes=shapes)


btn_mask.on_click(on_mask_clicked)
btn_trim.on_click(on_trim_clicked)
btn_reset_view.on_click(on_reset_view_clicked)
btn_reset_edits.on_click(on_reset_edits_clicked)
btn_export_fasta.on_click(on_export_fasta)
btn_export_fastq.on_click(on_export_fastq)

# Initial draw
update_start_slider_bounds()

fig.layout.template = "plotly_white"
ui = widgets.VBox([
    widgets.HBox([start_slider, win_size, pad_slider]),
    widgets.HBox([qmask_slider, btn_mask, trim_minq_slider, trim_window_slider, trim_minlen_slider, btn_trim]),
    widgets.HBox([btn_reset_view, btn_reset_edits, export_name, btn_export_fasta, btn_export_fastq]),
    status,
    fig
],
layout=widgets.Layout(width="100%"))

fig.layout.margin = dict(l=40, r=20, t=50, b=40)
fig.layout.legend = dict(orientation="h")

display(ui)

# ✅ Render AFTER display so Plotly knows the final container width
render()
#force_relayout()
wire_click_handler()


```
